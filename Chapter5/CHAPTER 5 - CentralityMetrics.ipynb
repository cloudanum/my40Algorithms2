```python
import networkx as nx
```

```python
import matplotlib.pyplot as plt
```

```python
vertices = range(1,10)
```

```python
edges = [(7,2), (2,3), (7,4), (4,5), (7,3), (7,5), (1,6),(1,7),(2,8),(2,9)]
```

```python
G = nx.Graph()
```

```python
G.add_nodes_from(vertices)
G.add_edges_from(edges)
```

```python
nx.draw(G, with_labels=True,node_color='y',node_size=800)
```

```python
nx.degree_centrality(G)
```

```python
nx.betweenness_centrality(G)
```

```python
nx.closeness_centrality(G)
```

```python
centrality = nx.eigenvector_centrality(G)
sorted((v, '{:0.2f}'.format(c)) for v, c in centrality.items())
```

```python
pos=nx.spring_layout(G)
nx.draw_networkx_nodes(G,pos,nodelist=[1,4,3,8,9],with_labels=True,node_color='g',node_size=1300)
nx.draw_networkx_nodes(G,pos, nodelist=[2,5,6,7],with_labels=True,node_color='r',node_size=1300)
nx.draw_networkx_edges(G,pos,edges,width=3,alpha=0.5,edge_color='b')
labels={}

labels[1]=r'1 NF'
labels[2]=r'2 F'
labels[3]=r'3 NF'
labels[4]=r'4 NF'
labels[5]=r'5 F'
labels[6]=r'6 F'
labels[7]=r'7 F'
labels[8]=r'8 NF'
labels[9]=r'9 NF'
nx.draw_networkx_labels(G,pos,labels,font_size=16)

#nx.draw_networkx_labels(G,pos,labels,font_size=16)
#nx.draw_networkx_edge_labels(G,pos,edge_labels={(7,2):'0.80', (2,3):'0.80', (7,4):'0.80', (4,5):'0.80', (7,3):'0.80', (7,5):'0.80', (1,6):'0.80',(1,7):'0.80',(2,8):'0.80',(2,9):'0.80'},font_color='r')
```
