```python
#Import the required Libraries
import numpy as np

import pandas as pd

import matplotlib.pyplot as plt
```

```python
#Generate the sampte data x and y using numpy.Linespace and np.square function respectivety
x = np.linspace(start = -10, stop = 10, num = 20, endpoint = True)
y = np.square(x-2)

plt.plot(x,y)

plt.xlabel('x', color = 'blue')

plt.ylabel('y', color = 'blue')

plt.show()
```

```python
#Gradient of y with respect to x

#For function y - np. square(x-2) the gradient is as beLow
gradient = lambda x: 2*(x-2)
```

```python
#Initial value of x
current_x = 1.0
```

```python
#This is precision,that tells us when to stop the algorithm
precision = 0.0001
```

```python
#maximum number of iterations
maximum_iterations = 1000
```

```python
# current interation

current_iteration = 0
```

```python
#Learning Rate

learning_rate = 0.01
```

```python
# previous step size - (currentx - previousx)

previous_step_size =0.5
```

```python
while previous_step_size > precision and current_iteration < maximum_iterations:
    previous_x = current_x
    current_x = current_x - learning_rate * gradient(previous_x)
    previous_step_size = abs(current_x - previous_x)
    current_iteration = current_iteration + 1
    print("Iteration ”, current iteration /‘Value of x is",current_x)
print(" The local minimum found at", current_x)
```
